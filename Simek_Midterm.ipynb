```python
import math
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
AAPL = pd.read_csv("AAPL.csv")
ALK = pd.read_csv("ALK.csv")
IPG = pd.read_csv("IPG.csv")
SEDG = pd.read_csv("SEDG.csv")
```

```python
aapl_open_close = AAPL["Open"] == AAPL["Close"]
pd_aapl_open_close = pd.DataFrame(data= aapl_open_close)
```

```python
aaplmax = AAPL["Close"].max()
alkmax = ALK["Close"].max()
ipgmax = IPG["Close"].max()
sedgmax = SEDG["Close"].max()

aaplmin = AAPL["Close"].min()
alkmin = ALK["Close"].min()
ipgmin = IPG["Close"].min()
sedgmin = SEDG["Close"].min()
```

```python
AAPL['Log Returns'] = np.log(AAPL['Close'] / AAPL['Close'].shift(1))
aaplmin1 = AAPL["Log Returns"].min()
aaplmax1 = AAPL["Log Returns"].max()
aaplmean1 = AAPL["Log Returns"].mean()
aaplmed1 = AAPL["Log Returns"].median()

ALK['Log Returns'] = np.log(ALK['Close'] / ALK['Close'].shift(1))
alkmin1 = ALK["Log Returns"].min()
alkmax1 = ALK["Log Returns"].max()
alkmean1 = ALK["Log Returns"].mean()
alkmed1 = ALK["Log Returns"].median()

IPG['Log Returns'] = np.log(IPG['Close'] / IPG['Close'].shift(1))
ipgmin1 = IPG["Log Returns"].min()
ipgmax1 = IPG["Log Returns"].max()
ipgmean1 = IPG["Log Returns"].mean()
ipgmed1 = IPG["Log Returns"].median()

SEDG['Log Returns'] = np.log(SEDG['Close'] / SEDG['Close'].shift(1))
sedgmin1 = SEDG["Log Returns"].min()
sedgmax1 = SEDG["Log Returns"].max()
sedgmean1 = SEDG["Log Returns"].mean()
sedgmed1 = SEDG["Log Returns"].median()
```

```python
AAPL_DATE = AAPL.set_index("Date")
appl_gain = AAPL_DATE["Log Returns"].idxmax()
appl_loss = AAPL_DATE["Log Returns"].idxmin()

ALK_DATE = ALK.set_index("Date")
alk_gain = ALK_DATE["Log Returns"].idxmax()
alk_loss = ALK_DATE["Log Returns"].idxmin()

IPG_DATE = IPG.set_index("Date")
ipg_gain = IPG_DATE["Log Returns"].idxmax()
ipg_loss = IPG_DATE["Log Returns"].idxmin()

SEDG_DATE = SEDG.set_index("Date")
sedg_gain = SEDG_DATE["Log Returns"].idxmax()
sedg_loss = SEDG_DATE["Log Returns"].idxmin()
```

```python
AAPL["Volume class"] = "zz"
AAPL.loc[AAPL['Volume'] <= (AAPL["Volume"]*0.25), 'Volume class'] = 'Low'
AAPL.loc[(AAPL['Volume'] > (AAPL["Volume"]*0.25)) & (AAPL['Volume'] <= (AAPL["Volume"]*0.5)), 'Volume class'] = 'Medium'
AAPL.loc[(AAPL['Volume'] > (AAPL["Volume"]*0.5)) & (AAPL['Volume'] <= (AAPL["Volume"]*0.75)), 'Volume class'] = 'Medium'
AAPL.loc[AAPL['Volume'] >= (AAPL["Volume"]*0.75), 'Volume class'] = 'Very High'
```

```python
AAPL.plot(kind= "line", x= "Date", y= "Log Returns", color = "blue")
ALK.plot(kind= "line", x= "Date", y= "Log Returns", color = "red")
IPG.plot(kind= "line", x= "Date", y= "Log Returns", color = "yellow")
SEDG.plot(kind= "line", x= "Date", y= "Log Returns", color = "green")
plt.show()
```

```python
AAPL.plot(kind= "hist", x= "Date", y= "Log Returns", color = "blue")
ALK.plot(kind= "hist", x= "Date", y= "Log Returns", color = "red")
IPG.plot(kind= "hist", x= "Date", y= "Log Returns", color = "yellow")
SEDG.plot(kind= "hist", x= "Date", y= "Log Returns", color = "green")
plt.show()
```
